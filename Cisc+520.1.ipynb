```python
import numpy as np
```

```python
#read in data
source_file="C:/Users/atan/Desktop/All/1 study/1 HU/g.520/voltage_3.csv"
data=np.genfromtxt(source_file,dtype="float32", delimiter=","
,skip_header=1)
data
```

```python
means=np.empty(8,"float32")
stds=np.empty(8,"float32")

for i in range(8):
    means[i]=np.average(data[:,i])
    stds[i]=np.std(data[:,i])
    
```

```python
means
stds
```

```python
#coloumn 1
data[:,1]
```

```python
avg=np.average(data[:,1])
sd=np.std(data[:,1])
avg
sd
```

```python
#includ column 1 to 6, only 6 columns
reduced_data=data[:,1:7]
reduced_data
```

```python
#normalize data from -1 to 1
normalized_data=(reduced_data)/10
print(normalized_data)
norm_means=means/10
norm_stds=stds/10
norm_means
norm_stds
```

```python
#plot
from matplotlib import pyplot as plt
```

```python
#force the number of rows to the list
time=list(range(len(normalized_data[:,0])))
print(time)
```

```python
#time vs voltage
#1 line plot
plt.plot(time,normalized_data[:,0])
plt.show()
```

```python
# scatter plot
plt.scatter(time,normalized_data[:,0])
plt.title("volt vs time, strain gauge 0")
plt.xlabel("time (ms)")
plt.ylabel("volt")
plt.show()
```

```python
#hist and normal prob distribution
import matplotlib.mlab as mlab
import scipy.stats as stats
n, bins, patches=plt.hist(normalized_data[:,0],200,normed=1)

expected=mlab.normpdf(bins, norm_means[1], norm_stds[1])
plt.plot(bins, expected,"p--")

plt.title("histogram of volt")
plt.xlabel("value Volt")
plt.ylabel("Freq")
plt.show()
```

```python
#qq plot 
# not normal, skewed below first part
column=0
stats.probplot(normalized_data[:, column], dist="norm",plot=plt)
plt.title("example qq plot1")
plt.show()
```

```python
#normal qq plot
column=1
stats.probplot(normalized_data[:, column], dist="norm",plot=plt)
plt.title("example qq plot2")
plt.show()
```

```python
#linspace
vals= np.linspace(0,1,1000)
def f(x):
    return x**3 +np.sin(x)
plt.scatter(vals,f(vals))
plt.show()
```

```python
#noisy data 
#systematic error: error due to data collection, miss read the scale, can be reduced
#indeterminate and random error: statistical error, can not be reudced. (noisy)
#able to fitler out noise under some circumstance
#white noise : perfectly normal distributed.
#other noise: fourier transformation to detect noise type (white, pink, blue)
plt.scatter(time, normalized_data[:,1], marker="x")
plt.ylim=[0.045,0.055]
plt.show()
```

```python
#white noise: almost a straight line
#blue noise: baised to higher noise
#signal data: big spiky noise around 60hz, AC hz, due to electric
plt.phase_spectrum(normalized_data[:,1], Fs=1000)
plt.title("blue noise example")
plt.show()
```
